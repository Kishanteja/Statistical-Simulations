```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Start', 'Revenue', 'Costs', 'Taxes', 'Profit']
values = [0, 400, -250, -50, 100]
cum_values = np.cumsum(values)
cum_values = np.insert(cum_values, 0, 0)[:-1]
fig, ax = plt.subplots(figsize=(10, 6))
colors = ['C0' if v >= 0 else 'C1' for v in values]
for i, (val, color) in enumerate(zip(values, colors)):
    ax.bar(categories[i], val, color=color, edgecolor='black', bottom=cum_values[i])
    height = val
    bottom = cum_values[i]
    yval = height + bottom
    ax.text(categories[i], yval, f'{yval:.0f}', ha='center', va='bottom')
plt.title('Waterfall Plot of Profit Changes')
plt.ylabel('Value')
plt.show()
```
