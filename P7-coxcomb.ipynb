```python
import matplotlib.pyplot as plt
import numpy as np

sections = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
sub_sections = ['Consumer', 'Corporate', 'Home Office']
data = {
    'Consumer': [50, 70, 60, 80, 90, 75, 85, 65, 55, 45, 60, 70],
    'Corporate': [80, 100, 90, 110, 105, 95, 115, 85, 75, 65, 90, 100],
    'Home Office': [65, 85, 75, 95, 85, 80, 90, 70, 65, 55, 70, 80]
}
colors = ['#FF0000', '#98FF98', '#6A5ACD']
num_vars = len(sections)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(polar=True))
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
width = 2 * np.pi / num_vars
for idx, sub_section in enumerate(sub_sections):
    values = data[sub_section]
    ax.bar(angles, values, width=width, bottom=0.0, color=colors[idx], alpha=0.7, label=sub_section)
ax.set_xticks(angles)
ax.set_xticklabels(sections)
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show()
```
