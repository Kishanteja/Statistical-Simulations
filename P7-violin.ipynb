```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

np.random.seed(10)
A = np.random.normal(70, 10, 1000)
B = np.random.normal(65, 15, 1000)
data = pd.DataFrame({
    'value': np.concatenate([A, B]),
    'data': ['A']*1000 + ['B']*1000
})

plt.figure(figsize=(10, 6))
sns.violinplot(x='data', y='value', data=data)
plt.title('violin plot')
plt.show()
```
