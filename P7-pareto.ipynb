```python
import matplotlib.pyplot as plt
import numpy as np

data = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
value = [40, 15, 10, 5, 30, 20, 25, 10, 8, 12]
sorted_data = sorted(zip(data, value), key=lambda x: x[1], reverse=True)
data, value = zip(*sorted_data)
cumulative_sum = np.cumsum(value)
cumulative_percentage = cumulative_sum / cumulative_sum[-1] * 100
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.bar(data, value, color='C0')
ax1.set_ylabel('value')
ax1.set_title('pareto chart')
ax2 = ax1.twinx()
ax2.plot(data, cumulative_percentage, color='C1', marker='o', linestyle='-', linewidth=2)
ax2.set_ylabel('Cumulative Percentage')
plt.show()
```
